```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#upload my dataset
df = pd.read_csv('dataset.csv')
df.head(100)
```

```python
#DATA PREPROCESSING
#delete column name 'id' as we dont need it as a feature for our model
df = df.drop('id', axis=1)
#MEAN IMPUTATION
#adding mean values to a missing value in a column
df['LSAS'] = df['LSAS'].fillna(df['LSAS'].median())
df.head()
```

```python
print(df['LSAS'].median())
```

```python
#FEATURE ENGINEERING
#FEATURE SCALING -> MIN-MAX NORMALIZATION to column names 'Age', 'EducationLevel', 'Occupation', 'ATF', 'AEF'. 'TKF', 'CMT', 'DEF', 'SMF'. 'ERF', 'DAF', 'SPIN'.'LSAS'
'''
https://towardsdatascience.com/everything-you-need-to-know-about-min-max-normalization-in-python-b79592732b79

Variables that are measured at different scales do not contribute equally to the model fitting & model learned function 
and might end up creating a bias. Thus, to deal with this potential problem feature-wise normalization such as 
MinMax Scaling is usually used prior to model fitting.

But we should use standardization because it does not concern itself with outliers.

1. https://stats.stackexchange.com/questions/10289/whats-the-difference-between-normalization-and-standardization
2. https://www.geeksforgeeks.org/normalization-vs-standardization/

'''
# df['Age']=(df['Age']-df['Age'].min())/(df['Age'].max()-df['Age'].min())
# df['EducationLevel']=(df['EducationLevel']-df['EducationLevel'].min())/(df['EducationLevel'].max()-df['EducationLevel'].min())
# df['Occupation']=(df['Occupation']-df['Occupation'].min())/(df['Occupation'].max()-df['Occupation'].min())
# df['ATF']=(df['ATF']-df['ATF'].min())/(df['ATF'].max()-df['ATF'].min())
# df['EAF']=(df['EAF']-df['EAF'].min())/(df['EAF'].max()-df['EAF'].min())
# df['TKF']=(df['TKF']-df['TKF'].min())/(df['TKF'].max()-df['TKF'].min())
# df['CMT']=(df['CMT']-df['CMT'].min())/(df['CMT'].max()-df['CMT'].min())
# df['DEF']=(df['DEF']-df['DEF'].min())/(df['DEF'].max()-df['DEF'].min())
# df['SMF']=(df['SMF']-df['SMF'].min())/(df['SMF'].max()-df['SMF'].min())
# df['ERF']=(df['ERF']-df['ERF'].min())/(df['ERF'].max()-df['ERF'].min())
# df['DAF']=(df['DAF']-df['DAF'].min())/(df['DAF'].max()-df['DAF'].min())
# df['SPIN']=(df['SPIN']-df['SPIN'].min())/(df['SPIN'].max()-df['SPIN'].min())
# df['LSAS']=(df['LSAS']-df['LSAS'].min())/(df['LSAS'].max()-df['LSAS'].min())

'''
The code below is similar to this one:
    
    from sklearn.preprocessing import StandardScaler
    scaler = StandardScaler()
    scaled = scaler.fit_transform(df[ ['Age', 'EducationLevel', 'Occupation', 'ATF', 'EAF', 'TKF', 'CMT', 'DEF', 'SMF', 'ERF', 'DAF', 'SPIN', 'LSAS'] ])
    
However, the scaled would only have the columns that were standardized, how about those columns that were
not standardized? That is why, we have to manually do the Standardization using the formula X`= (X − μ) / σ.
'''
df['Age'] = (df['Age'] - df['Age'].mean()) / np.std(df['Age'])
df['EducationLevel'] = (df['EducationLevel'] - df['EducationLevel'].mean()) / np.std(df['EducationLevel'])
df['Occupation'] = (df['Occupation'] - df['Occupation'].mean()) / np.std(df['Occupation'])
df['ATF'] = (df['ATF'] - df['ATF'].mean()) / np.std(df['ATF'])
df['EAF'] = (df['EAF'] - df['EAF'].mean()) / np.std(df['EAF'])
df['TKF'] = (df['TKF'] - df['TKF'].mean()) / np.std(df['TKF'])
df['CMT'] = (df['CMT'] - df['CMT'].mean()) / np.std(df['CMT'])
df['DEF'] = (df['DEF'] - df['DEF'].mean()) / np.std(df['DEF'])
df['SMF'] = (df['SMF'] - df['SMF'].mean()) / np.std(df['SMF'])
df['ERF'] = (df['ERF'] - df['ERF'].mean()) / np.std(df['ERF'])
df['DAF'] = (df['DAF'] - df['DAF'].mean()) / np.std(df['DAF'])
df['SPIN'] = (df['SPIN'] - df['SPIN'].mean()) / np.std(df['SPIN'])
df['LSAS'] = (df['LSAS'] - df['LSAS'].mean()) / np.std(df['LSAS'])


df.head(15)
```

```python
#FEATURE SELECTION
#FINDING THE BEST FEATURE  To IMPROVE THE ACCURACY
#Univariate Selection
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
'''
https://stackoverflow.com/questions/25792012/feature-selection-using-scikit-learn
computes ANOVA f-value
'''
from sklearn.feature_selection import f_classif


X = df.loc[:, df.columns != 'hasSAD']  #features

y = df.loc[:, df.columns == 'hasSAD']  #features    #label


#apply SelectKBest class to extract top 10 best features
# total features except the dependent / 2
n_features = int(X.shape[1] / 2)
bestfeatures = SelectKBest(score_func=f_classif, k = n_features)
fit = bestfeatures.fit(X, np.asarray(y['hasSAD']))
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
#concat two dataframes for better visualization 
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Columns','Score']  #naming the dataframe columns
print(featureScores.nlargest(n_features,'Score'))  #print 10 best features

```

```python
#Feature Importance
from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt
model = ExtraTreesClassifier()
model.fit(X,y)
print(model.feature_importances_) #use inbuilt class feature_importances of tree based classifiers
#plot graph of feature importances for better visualization
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(15).plot(kind='barh')
plt.show()
```

```python
import seaborn as sns

#get correlations of each features in dataset
corrmat = df.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(20,20))
#plot heat map
g=sns.heatmap(df[top_corr_features].corr(),annot=True,cmap="RdYlGn")
```

```python
#CREATE THE FINAL DATASET WITH THE BEST CHOSEN FEATURES ACCORDING TO THE RESULT IN FEATURE SELECTION
final_data = df[['SPIN','LSAS','TKF','ERF','CMT','ATF','SMF','DAF','DEF','EAF','hasSAD']].copy()
final_data.head(200)
```

```python
#make a list for the features and labels
#create and train the model without usign kfold
import sklearn
from sklearn.utils import shuffle
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix

spin = np.array(final_data['SPIN'])
lsas = np.array(final_data['LSAS'])
tkf = np.array(final_data['TKF'])
erf = np.array(final_data['ERF'])
cmt = np.array(final_data['CMT'])
atf = np.array(final_data['ATF'])
smf = np.array(final_data['SMF'])
daf = np.array(final_data['DAF'])
deF = np.array(final_data['DEF'])
eaf = np.array(final_data['EAF'])
hasSAD = np.array(final_data['hasSAD'])
#features
x = np.array(list(zip(spin,lsas,tkf,erf,cmt,atf,smf,daf,deF,eaf)))
#labels
y = np.array(hasSAD)
#normal splitting
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x,y,test_size = 0.2)

knn_model = KNeighborsClassifier(n_neighbors = 5)
knn_model.fit(x_train,y_train)
acc = knn_model.score(x_test,y_test)
prediction = knn_model.predict(x_test)
print(acc)
cm = confusion_matrix(y_test,prediction)
print(cm)
```

```python
#Implementing K-Fold cross-validation
from sklearn.model_selection import KFold 
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier

k = 10
kf = KFold(n_splits=k, random_state=None)
clf = KNeighborsClassifier(n_neighbors = 7)
# clf = DecisionTreeClassifier(splitter = 'random', random_state = 42)
acc_score = []

for train_index , test_index in kf.split(x):
    X_train , X_test = x[train_index], x[test_index]
    y_train , y_test = y[train_index] , y[test_index]

    clf.fit(X_train,y_train)
    pred_values = clf.predict(X_test)
     
    acc = accuracy_score(pred_values , y_test)
    acc_score.append(acc)
     
avg_acc_score = sum(acc_score)/k
 
print('accuracy of each fold - {}'.format(acc_score))
print('Avg accuracy : {}'.format(avg_acc_score))


accuracy = metrics.make_scorer(metrics.accuracy_score)

# clf = DecisionTreeClassifier(splitter = 'random', random_state = 42)
# print(cross_val_score(clf, x, y, scoring = accuracy, cv = kf).mean())



'''
0.886147



'''
```
